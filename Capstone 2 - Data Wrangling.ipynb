```python
#Count (using `.sum()`) the number of missing values (`.isnull()`) in each column of 
#walmart_data as well as the percentages (using `.mean()` instead of `.sum()`).
#Order them (increasing or decreasing) using sort_values
#Call `pd.concat` to present these in a single table (DataFrame) with the helpful column names 'count' and '%'
missing = pd.concat([walmart_data.isnull().sum(), 100 * walmart_data.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count', ascending = False)
```

```python
# Good! Looks like there are zero missing values in the data.
# Now I want to know how many stores/how much data is available per store.
# I can tabulate the distinct values along with their respective frequencies using value_counts()

walmart_data['Store'].value_counts()
```

```python
# Per the above, it looks like we have 45 stores, and each store has 143 entries of data
# I want to confirm the number of unique stores.
# I will use the `nunique` method

walmart_data[['Store']].nunique()
```

```python
# I was right! There are 45 different stores, numbered from 1 to 45
# I want to start looking at the data a little closer
# For each store, I want to calculate average weekly sale and sort by the average

store_weeklysalesprice_means = walmart_data.groupby('Store')[['Weekly_Sales']].mean()
store_weeklysalesprice_means.head()
```

```python
# Now I want to look at the above in a bar graph

(store_weeklysalesprice_means.reindex(index=store_weeklysalesprice_means.mean(axis=1)
    .sort_values(ascending=False)
    .index)
    .plot(kind='barh', figsize=(10, 10), title='Average Weekly Sales by Store Number'))
plt.xlabel('Weekly Sales ($)');
```

```python
# Now, for each store, I want to calculate average temperature and sort by the average

store_temperature_means = walmart_data.groupby('Store')[['Temperature']].mean()
store_temperature_means.head()
```

```python
# Now I want to look at the above in a bar graph

(store_temperature_means.reindex(index=store_temperature_means.mean(axis=1)
    .sort_values(ascending=False)
    .index)
    .plot(kind='barh', figsize=(10, 10), title='Average Temperature by Store Number'))
plt.xlabel('Temperature (F)');
```
